```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from lxml import etree

opts = Options()
opts.add_argument("--headless=new")
opts.add_argument("--disable-gpu")
opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64)")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=opts)

try:
    url = "https://y.qq.com/n/ryqq/toplist/4"
    print("Opening:", url)
    driver.get(url)

    wait = WebDriverWait(driver, 15)
    wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="app"]/div/div[2]/div[2]/div[3]/ul[2]/li[1]')))

    page_src = driver.page_source
    print("Page source length:", len(page_src))

finally:
    driver.quit()
    print("Driver quit")

# 使用 lxml 解析
dom = etree.HTML(page_src)

# XPath 选择器：和之前用过的路径一致，取每个条目的第 5 个 div（时长）
xpath = '//*[@id="app"]/div/div[2]/div[2]/div[3]/ul[2]/li/div/div[5]/text()'
raw_durations = dom.xpath(xpath)

# 清理并打印
durations = [d.strip() for d in raw_durations if d.strip()]
print("Found durations count:", len(durations))
print("durations:", durations)
```

```python
from lxml import etree

dom = etree.HTML(page_src)  # 假设之前单元格已生成 page_src
# 选择目标 div（示例：每个条目的根 div），请按实际结构调整 XPath
xpath_items = '//*[@id="app"]/div/div[2]/div[2]/div[3]/ul[2]/li/div'
items = dom.xpath(xpath_items)
texts = items[1].xpath('.//text()')
for i, node in enumerate(items, start=1):
    texts = node.xpath('.//text()')                 # 获取节点及子节点所有文本节点
    cleaned = ' '.join(t.strip() for t in texts if t.strip())  # 清理并合并
    print(f'Item {i}:', cleaned)
```

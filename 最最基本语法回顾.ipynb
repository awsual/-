 单行注释

Python编程语言的单行注释常以#开头，单行注释可以作为单独的一行放在被注释代码行之上，也可以放在语句或者表达式之后。

```python
print ("hello world!");#您好，世界
```

```python
# 1.1 基本数据类型定义
int_var = 10  # 整数
float_var = 3.14  # 浮点数
str_var = "Python基础回顾"  # 字符串
bool_var = True  # 布尔值
none_var = None  # 空值

# Jupyter自动打印最后一行的变量（多个变量用逗号分隔）
int_var, float_var, str_var, bool_var, none_var
```

```python
# 1.2 复合数据类型定义
list_var = [1, 2, "a", [3, 4]]  # 列表（有序、可修改）
tuple_var = (1, 2, "b")  # 元组（有序、不可修改）
dict_var = {"name": "小明", "age": 20}  # 字典（键值对）
set_var = {1, 2, 3, 3}  # 集合（无序、去重）

# 自动打印复合类型变量
list_var, tuple_var, dict_var, set_var
```

```python

# 1.3 数据类型查看与转换



# Jupyter自动打印最后一行的变量（多个变量用逗号分隔）
int_var, float_var, str_var, bool_var, none_var
print("str_var的类型：", type(str_var))
print("float_var转整数：", int(float_var))
print("tuple_var转列表：", list(tuple_var))
```

```python
# 2. 运算符示例
a = 10
b = 3

# 算术运算符
print("加法：", a + b)
print("整除：", a // b)
print("幂运算：", a **b)

# 比较与逻辑运算符
print("a > b吗？", a > b)
print("a>5且b<5？", a > 5 and b < 5)

# 成员运算符（最后一行自动打印）
"a" in list_var, "name" in dict_var  # 列表中是否有"a"，字典中是否有"name"键
```

```python
# 3. 条件判断示例
score = 85

if score >= 90:
    print("优秀")
elif score >= 80:
    print("良好")
elif score >= 60:
    print("及格")
else:
    print("不及格")

# 三元运算符（最后一行自动打印）
"及格" if score >= 60 else "不及格"
```

```python
# 4.1 for循环示例
print("遍历列表：")
for num in [1, 2, 3, 4]:
    print(num * 2, end=" ")

print("\n\n遍历字典：")
for key, value in dict_var.items():
    print(f"{key}: {value}")

# range()函数（最后一行自动打印遍历结果）
[i for i in range(3, 10, 2)]  # 用列表推导式收集结果，方便自动打印
```

```python
# 4.2 while循环
count = 0
print("while循环：")
while count < 3:
    print(f"循环第{count+1}次")
    count += 1

# 4.3 循环控制（break/continue）
print("\n循环控制：")
result = []
for i in range(5):
    if i == 2:
        continue  # 跳过i=2
    if i == 4:
        break  # 终止循环
    result.append(i)

# 自动打印结果
result
```

```python
# 5.1 基础函数
def add(a, b):
    """两数相加"""
    return a + b

# 5.2 带默认参数的函数
def greet(name="陌生人"):
    print(f"你好，{name}！")

# 调用函数（最后一行自动打印结果）
greet("Python学习者"), add(3, 5)  # 函数调用结果（greet无返回值，显示None）
```

```python
# 6. 列表推导式示例
# 1-10的平方
square_list = [i**2 for i in range(1, 11)]
# 筛选偶数并乘2
even_double = [x*2 for x in range(1, 20) if x % 2 == 0]

# 自动打印两个列表
square_list, even_double
```

```python
# 7. 异常处理示例
try:
    result = 10 / 0  # 会触发除零异常
except ZeroDivisionError:
    print("错误：除数不能为0！")
except Exception as e:
    print(f"未知错误：{e}")
else:
    print(f"结果：{result}")
finally:
    print("执行结束")
```

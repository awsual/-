替换数字

```python
import re

result1 = re.sub(r"\d+", "", "runoob123google456")
result2 = re.sub(r"\d+", "", "run88oob123google456")

print(result1)
print(result2)
```

```python
import re

pattern = re.compile(r'\d+')  # 查找数字
result1 = pattern.findall('runoob 123 google 456')
result2 = pattern.findall('run88oob123google456', 0, 10)

print(result1)
print(result2)
```

re.searh扫描整个字符串并返回第一个成功的匹配。


```python
import re

print(re.search('www', 'www.runoob.com'))
print(re.search('www', 'www.runoob.com').span())
print(re.search('com', 'www.runoob.com').span())
```

从一段文本中提取所有的URL 链接

```python
import re

def extract_urls(text):
    pattern = r'https?://[^\s]+'  # 使用原始字符串和标准直引号
    return re.findall(pattern, text)

text = 'Visit my website at https://www.example.com or check outhttp://blog.example.net'
print(extract_urls(text))
```
